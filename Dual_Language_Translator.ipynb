```python
import pandas as pd
```

```python
data = pd.read_csv('/content/dataset1.csv', sep=None, engine='python', on_bad_lines='skip')

print(data)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier

le_en = LabelEncoder()
le_fr = LabelEncoder()
le_hi = LabelEncoder()

data['English'] = le_en.fit_transform(data['English'])
data['French'] = le_fr.fit_transform(data['French'])
data['Hindi'] = le_hi.fit_transform(data['Hindi'])

X = data[['English']]
y_fr = data['French']
y_hi = data['Hindi']

model_fr = DecisionTreeClassifier()
model_hi = DecisionTreeClassifier()

model_fr.fit(X, y_fr)
model_hi.fit(X, y_hi)
```

```python
def translate_word(word):

    if len(word) < 10:
        return "Upload Again"

    try:
        word_encoded = le_en.transform([word])

        fr_pred = model_fr.predict([word_encoded])
        hi_pred = model_hi.predict([word_encoded])

        french = le_fr.inverse_transform(fr_pred)
        hindi = le_hi.inverse_transform(hi_pred)

        return french[0], hindi[0]

    except:
        return "Word not found in dataset"
```

```python
import ipywidgets as widgets
from IPython.display import display
```

```python
english_input = widgets.Text(description='Enter English Word:')
output_area = widgets.Output()

print("Input and output widgets created.")
```

```python
from IPython.display import clear_output

def on_translate_button_clicked(button):
    with output_area:
        clear_output()
        word = english_input.value
        result = translate_word(word)

        if result == "Upload Again":
            print("Upload Again")
        elif result == "Word not found in dataset":
            print("Word not found")
        else:
            print(f"French: {result[0]}\nHindi: {result[1]}")

print("Translation function 'on_translate_button_clicked' defined.")
```

```python
translate_button = widgets.Button(description='Translate')
translate_button.on_click(on_translate_button_clicked)

display(english_input, translate_button, output_area)

print("GUI displayed.")
```
